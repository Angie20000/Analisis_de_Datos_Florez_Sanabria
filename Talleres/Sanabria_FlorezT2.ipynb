### 2. Crear el Dataset
Ejecuta la siguiente celda para crear los DataFrames que usaremos en los ejercicios.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set_theme(style='whitegrid')
np.random.seed(42)

# Dataset 1: Ventas mensuales
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas_2022 = [45, 52, 48, 65, 70, 85, 90, 88, 75, 80, 95, 110]
ventas_2023 = [50, 58, 55, 72, 78, 92, 98, 95, 82, 88, 105, 120]

df_ventas = pd.DataFrame({
    'Mes': meses,
    'Ventas_2022': ventas_2022,
    'Ventas_2023': ventas_2023
})

# Dataset 2: Datos de empleados
n_empleados = 150
df_empleados = pd.DataFrame({
    'Edad': np.random.normal(35, 8, n_empleados).astype(int).clip(22, 65),
    'Salario': np.random.normal(50000, 15000, n_empleados).clip(25000, 120000),
    'Años_Experiencia': np.random.normal(8, 4, n_empleados).clip(0, 30),
    'Departamento': np.random.choice(['Ventas', 'IT', 'Marketing', 'RRHH'], n_empleados),
    'Satisfaccion': np.random.randint(1, 11, n_empleados)
})

# Dataset 3: Productos
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares']
unidades_vendidas = [120, 450, 380, 210, 290]

df_productos = pd.DataFrame({
    'Producto': productos,
    'Unidades': unidades_vendidas
})

print('Datasets creados exitosamente')
print(f'\ndf_ventas: {df_ventas.shape}')
print(f'df_empleados: {df_empleados.shape}')
print(f'df_productos: {df_productos.shape}')
```

### 3. Gráfico de Líneas Simple
Crea un gráfico de líneas que muestre las ventas de 2022 (`Ventas_2022`) a lo largo de los meses. Usa `plt.plot()` y añade título y etiquetas a los ejes.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2022'], color='#BB00FF')
plt.title('Ventas Mensuales de 2022')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.savefig('Grafico3.png', dpi=400)
plt.show()
```

De acuerdo a la gráfica anterior de ventas 2022 por meses, se evidencia lo siguiente:
- El promedio de ventas del 2022 fue de 77.5.
- Los meses de mayor auge son (julio, octubre, noviembre y diciembre) con un promedio de 92.5.
- Los meses de menores ventas fueron: Enero, febrero y marzo esto posiblemente debido a comienzo del año y temas escolares y postvacaciones.

```python
df_ventas['Ventas_2022'].describe().T
```

### 4. Gráfico de Líneas Múltiples
Crea un gráfico que muestre las ventas de 2022 y 2023 en la misma figura. Usa diferentes colores y añade una leyenda.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2022'], color='#BB00FF', label='2022')
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2023'], color='#00FFC4', label='2023')
plt.title('Ventas del 2022 vs 2023')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

De acuerdo a la gráfica anterior evidenciamos lo siguiente:

- El aumento de ventas ha generado un aumento en la dispersión de los datos, con una poca posibilidad de predicción de ventas.

```python
df_ventas[['Ventas_2022','Ventas_2023']].describe().T
```

### 12. Histograma con Seaborn (Histplot)
Crea un histograma de la distribución de salarios usando `sns.histplot()`. Añade una curva KDE (Kernel Density Estimate).

```python
# Tu código aquí
plt.figure(figsize=(10,6))
sns.histplot(df_empleados['Salario'], kde=True)
plt.title('Distribución de Salarios')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
```

El histograma revela cómo se distribuyen los sueldos entre los empleados:

¿Es simétrica? → Indica uniformidad y políticas salariales consistentes.

¿Está sesgada a la derecha? (lo más común) → Significa que la mayoría de empleados tiene salarios más bajos o medios, con pocos salarios muy altos.

¿Está sesgada a la izquierda? → Caso raro: mayoría gana salarios altos.

Esta lectura ayuda a entender la pirámide salarial interna.

### 14. Scatter Plot con Seaborn

### 15. Heatmap de Correlación
Crea un heatmap que muestre la matriz de correlación de las variables numéricas de `df_empleados`. Usa `sns.heatmap()` con `annot=True`.

```python
# Tu código aquí
plt.figure(figsize=(10,6))
correlacion_matrix = df_empleados[['Edad', 'Salario','Años_Experiencia','Satisfaccion']].corr(method='spearman')
sns.heatmap(correlacion_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```


segun el diagrama: identificar relaciones lineales entre variables numéricas de df_empleados.
Valor analítico:

Detecta patrones subyacentes que no son obvios a simple vista.

Facilita decisiones como: ¿qué variable predice mejor el salario?, ¿hay multicolinealidad?

El uso de annot=True agrega transparencia y precisión ejecutiva.

Conclusión: herramienta clave para diagnóstico preliminar y análisis exploratorio de datos (EDA).

### 16. Pair Plot

### 17. Subplots (Múltiples Gráficos)
Crea una figura con 2 subplots (1 fila, 2 columnas):
- Subplot 1: Histograma de edades
- Subplot 2: Histograma de salarios

Usa `plt.subplots()`.

```python
# Tu código aquí
fig, axes = plt.subplots(2, 2, figsize=(15, 5))

# Subplot 1: Edades
axes[0, 0].hist(df_empleados['Edad'], bins=15, color='pink')
axes[0, 0].set_title('Distribución de Edades')
axes[0, 0].set_xlabel('Edad')
axes[0, 0].set_ylabel('Frecuencia')

# Subplot 2: Salarios
axes[0, 1].hist(df_empleados['Salario'], bins=15, color='green')
axes[0, 1].set_title('Distribución de Salarios')
axes[0, 1].set_xlabel('Salario')
axes[0, 1].set_ylabel('Frecuencia')

# Subplot 3 (opcional, puedes ajustar el contenido)
axes[1, 0].hist(df_empleados['Edad'], bins=15)
axes[1, 0].set_title('Distribución de Edades')
axes[1, 0].set_xlabel('Edad')
axes[1, 0].set_ylabel('Frecuencia')

# Subplot 4 (opcional)
axes[1, 1].hist(df_empleados['Salario'], bins=15, color='orange')
axes[1, 1].set_title('Distribución de Salarios')
axes[1, 1].set_xlabel('Salario')
axes[1, 1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Objetivo técnico: presentar comparaciones visuales en una misma figura, optimizando el espacio y mejorando la narrativa.
Valor analítico:

Histograma de Edad: evidencia estructura demográfica.

Histograma de Salario: refleja dispersión, simetría y posibles brechas internas.

Uso de plt.subplots() profesionaliza la presentación y facilita reportes ejecutivos.

Conclusión: visibilidad clara y organizada para contrastar variables clave en un mismo lienzo.

### 19. Personalización Avanzada
Crea un scatter plot de `Años_Experiencia` vs `Salario` con las siguientes personalizaciones:
- Tamaño de figura: 10x6
- Color de puntos: verde
- Transparencia (alpha): 0.6
- Tamaño de puntos: 50
- Grid activado
- Título con tamaño de fuente 16
- Etiquetas de ejes con tamaño de fuente 12

```python
# Tu código aquí
plt.figure(figsize=(10, 6),)
plt.scatter(
    df_empleados['Años_Experiencia'],
    df_empleados['Salario'],
    color='purple',
    alpha=0.6,
    s=50
)

plt.grid(True)
plt.title('Experiencia vs Salario', fontsize=16)
plt.xlabel('Años de Experiencia', fontsize=12)
plt.ylabel('Salario', fontsize=12)
plt.show()
```

Objetivo técnico: correlacionar dos variables clave en RRHH / talento humano.
Valor analítico:

Mide la coherencia salarial en relación con la experiencia.

Permite detectar inequidades, empleados fuera de banda o talento senior subpagado.

Personalización mejora legibilidad y hace la gráfica “lista para comité”.

Conclusión: insumo de alto valor para auditoría salarial y planificación de talento.
